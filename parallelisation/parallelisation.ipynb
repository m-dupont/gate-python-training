```python
#do code here, do not override cell below
```

```python
def make_image(s):
    binx = np.linspace(-25,25 , 51)
    biny = np.linspace(-25,25 , 51)
    H, xedges, yedges = np.histogram2d(s['globalPosX'], s['globalPosY'], bins=(binx, biny))
    H = np.flip(H.T,0)
    return H
```

```python
ret = df.groupby('runID').apply(make_image)
stack = np.stack(ret)
```

## Threads

```python
import time
import threading
```

```python
def sleep_5():
    print("doing sleep_5 in ", threading.current_thread())
    time.sleep(5)
    print("doing sleep_5 done in ", threading.current_thread())
```

Execute the function

```python
%%time 
sleep_5()
```

## launch threads from python

```python
t = threading.Thread(target=sleep_5)
```

```python
t.run()
print("after run")
```

```python
t = threading.Thread(target=sleep_5)
t.start()
print("after start")
```

```python
t = threading.Thread(target=sleep_5)
t.start()
print("after start")
t.join()
print("after join")
```
